# f1 and p1:


```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def f1(x, y):
    return x ** 2 + y ** 2 + 2

x = np.linspace(-6, 6, 300)
y = np.linspace(-6, 6, 300)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(0,0,2)
ax.view_init(60, 35)
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.scatter(0,0,20)
```

```
sns.distplot(Z,hist=False, color='r')
plt.scatter(0,0)
```

$$f_1 = x^2 + y^2$$
$$p_1 = (0,0,2) $$

# f3 and p3:


```
def f3(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(85, 40)
ax.scatter(0,0,0)
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.scatter(0,0,10)
```

```
sns.distplot(Z,hist=False, color='r')
plt.scatter(0,0)
```

$$f_3 = sin(\sqrt{x^2 + y^2})$$
$$p_3 = (0,0,0)$$
